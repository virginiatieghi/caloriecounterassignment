```python
calories = {
   'Hamburger': 600,
   'Cheese Burger': 750,
   'Veggie Burger': 400,
   'Vegan Burger': 350,
   'Sweet Potatoes': 230,
   'Salad': 15,
   'Iced Tea': 70,
   'Lemonade': 90,
   }


x=0
for i in calories:
    #print(i in calories)
    #print(calories[i])  
    x = x + calories[i]

print(x)

#Create a calorie counter function

def calorie_counter(order):
    total_calories=0
    for item in order:
        total_calories = total_calories + calories[item]
    return total_calories


result = calorie_counter(["Hamburger", "Iced Tea"])
print(result)

print(calorie_counter(calories))

list_1= {
   'Hamburger': 600,
   'Cheese Burger': 750   }

print(calorie_counter(list_1))

result = calorie_counter(["Hamburger", "Iced Tea"])
print(result)

def calorie_counter_with_args(*order):
    x=0
    for i in order:
        x = x + calories[i]
    return x
print(calorie_counter(calories))

result = calorie_counter_with_args("Hamburger", "Iced Tea")
print(result)

combos = {
    "Cheesy Combo" : ["Cheese Burger", "Sweet Potatoes", "Lemonade"],
    "Veggie Combo" : ["Veggie Burger", "Sweet Potatoes", "Iced Tea"],
    "Vegan Combo" : ["Vegan Burger", "Salad", "Lemonade"],
}

#Handle combos

def combo_calories_counter (ordered_item):
    total_calories=0
    for i in ordered_item:
        if i in calories:
           total_calories= total_calories + calories[i]
        elif i in combos:
            combo_items= combos[i]
            total_calories= total_calories + sum(calories[item]for item in combo_items)
        else:
            print(f"{ordered_item} is not in the calorie database or is not a valid combo.")
    return total_calories

meal_order = ["Cheesy Combo", "Veggie Burger", "Iced Tea"]
total_calories = combo_calories_counter(meal_order)
print(f"Total calories for the meal order: {total_calories} calories")

#Handle errors

def combo_calories_counter2 (ordered_item):
    total_calories=0
    for i in ordered_item:
        try:
            if i in calories:
               total_calories= total_calories + calories[i]
            elif i in combos:
                combo_items= combos[i]
                total_calories= total_calories + sum(calories[item]for item in combo_items)
            else:
                raise KeyError
        except KeyError:
            print( print(f"The item '{ordered_item}' is not on the menu."))
    return total_calories

total_calories = combo_calories_counter2(meal_order)
print(f"Total calories for the meal order: {total_calories} calories")

#Use more complex data

meals_list = [
    {
        "id": "meal-1",
        "name": "hamburger",
        "calories": 600,
        "price": 5
    },
    {
        "id": "meal-2",
        "name": "cheese burger",
        "calories": 750,
        "price": 7
		},
    {
        "id": "meal-3",
        "name": "veggie burger",
        "calories": 400,
        "price": 6
		},
    {
        "id": "meal-4",
        "name": "vegan burger",
        "calories": 350,
        "price": 6
		},
    {
        "id": "meal-5",
        "name": "sweet potatoes",
        "calories": 230,
        "price": 3
		},
    {
        "id": "meal-6",
        "name": "salad",
        "calories": 15,
        "price": 4
		},
    {
        "id": "meal-7",
        "name": "iced tea",
        "calories": 70,
        "price": 2
		},
    {
        "id": "meal-8",
        "name": "lemonade",
        "calories": 90,
        "price": 2
		}
]

combos_list = [
    {
        "id": "combo-1",
        "name": "cheesy combo",
        "meals": ["meal-2", "meal-5", "meal-8"],
        "price": 11,
    },
    {
        "id": "combo-2",
        "name": "veggie combo",
        "meals": ["meal-3", "meal-5", "meal-7"],
        "price": 10,
    },
    {
        "id": "combo-3",
        "name": "vegan combo",
        "meals": ["meal-4", "meal-6", "meal-8"],
        "price": 10,
    }
]

meals_dict = {
	meals["id"]: meals
	for meals in meals_list
}
print(meals_dict)

combos_dict = {
	combos["id"]: combos
	for combos in combos_list
}
print(combos_dict)

#Create a price counter function

def calculate_total_price(ordered_items, meals_dict, combos_dict):
    total_price = 0

    for item in ordered_items:
        try:
            if item in meals_dict:
                total_price += meals_dict[item]["price"]
            elif item in combos_dict:
                total_price += combos_dict[item]["price"]
            else:
                raise KeyError
        except KeyError:
            print(f"{item} is not in the menu.")
    return total_price

ordered_items = ["meal-2", "combo-1", "meal-5", "meal-8"]
total_price = calculate_total_price(ordered_items, meals_dict, combos_dict)
print(f"Total price for the ordered items: ${total_price}")

#Store your data in json files

import json

with open("data/meals.json") as file:
	meals = json.load(file)

print(meals)

import json

with open("data/combos.json") as file:
	combos = json.load(file)

print(combos)






     
```
